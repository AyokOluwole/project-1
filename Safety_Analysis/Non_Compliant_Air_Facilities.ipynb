```python
import pandas as pd
```

```python
# Store filepath in a variable
air_facilities = "../Project/Resources/Raw_Data/2021_Facilities_Air_Emission_Non_compliant.csv"
```

```python
# Read the Data file with the pandas library
af_df = pd.read_csv(air_facilities, encoding_errors= 'replace')

af_df.head(3)
```

```python
organized_df = af_df[['District', 'Type of Exceedance']]

organized_df = organized_df.rename(columns={'District': 'City', 'Type of Exceedance': '# of Non Compliant Facilities for Air Emission'})

organized_df.head()
```

```python
#Group by Cities and sum the categories
grouped_df = organized_df.groupby(['City']).count()

grouped_df.sort_values(['# of Non Compliant Facilities for Air Emission'], ascending=False)
```

```python
#Save the final copy to CSV
grouped_df.to_csv("../Project/Resources/Cleaned_Data/Air_Emission_Non_Compliant_Facilities.csv")
```
