```python
import pandas as pd
```

```python
# Store filepath in a variable
infra_proj = "../Project/Resources/Raw_Data/Ont_infrastructure.csv"
```

```python
# Read the Data file with the pandas library
infra_proj_df = pd.read_csv(infra_proj)

infra_proj_df.head(3)
```

```python
#Keep the required Columns
organized_df = infra_proj_df[['Community', 'Category']]

organized_df = organized_df.rename(columns= {'Community': 'City', 'Category': '# of Planned Infrastructure Projects' })

organized_df.head()
```

```python
#Group by Cities and sum the categories
grouped_df = organized_df.groupby(['City']).count()

sorted_df = grouped_df.sort_values(['# of Planned Infrastructure Projects'], ascending=False)

sorted_df.head()
```

```python
#Save the final copy to CSV
sorted_df.to_csv("../Project/Resources/Cleaned_Data/Infrastructure_Projects.csv")
```
