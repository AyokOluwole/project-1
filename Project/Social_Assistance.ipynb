```python
import pandas as pd
```

```python
# Store filepath in a variable
social_assistance = "../Project/Resources/Raw_Data/Social_Assistance_and_Disability_Support_Program_Cases.csv"
```

```python
# Read the Data file with the pandas library
sa_df= pd.read_csv(social_assistance)

sa_df.head()
```

```python
#Remove the unnecessary columns
organized_df = sa_df[['CMA', 'Cases']]

organized_df = organized_df.rename(columns={'CMA': "City", 'Cases': "Social Assistance Programs Offered"})

organized_df.head()
```

```python
# Convert from String to Int
df = organized_df.astype({'Social Assistance Programs Offered':'int'})

grouped_df = df.groupby(['City']).sum()

grouped_df.sort_values(['Social Assistance Programs Offered'], ascending=False)
```

```python
#Save the final copy to CSV
grouped_df.to_csv("../Project/Resources/Cleaned_Data/Social_Assistance_Programs.csv")
```
