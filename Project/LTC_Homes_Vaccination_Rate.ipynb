```python
import pandas as pd
```

```python
# Store filepath in a variable
ltc_file = "../Project/Resources/Raw_Data/Long_Term_Care_Homes_Vaccination_Rate.csv"
```

```python
# Read the Data file with the pandas library
ltc_df = pd.read_csv(ltc_file)

ltc_df.head()
```

```python
#Drop Duplicates in LTC Home and keep the "Last Date Collected Data"

dropped_df = ltc_df.drop_duplicates(subset='LTC_Home', keep="last")

dropped_df.head()
```

```python
#Organize the df by keeping the required columns

organized_df = dropped_df[['City', '1st_dose_percentage_staff_vaccination_rate', '2nd_dose_percentage_staff_vaccination_rate']]

organized_df.head()
```

```python
#Group by Cities and get the average 
grouped_df = organized_df.groupby(['City']).mean()

grouped_df.head()
```

```python
#Sort by highest 2nd vaccination rate

grouped_df.sort_values(['2nd_dose_percentage_staff_vaccination_rate'], ascending=False)
```

```python
#Save the final copy to CSV

grouped_df.to_csv("../Project/Resources/Cleaned_Data/LTC_Staff_Vaccination_Rate.csv")
```
