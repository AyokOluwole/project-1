```python
import pandas as pd
```

```python
# Store filepath in a variable
uni_enrol_file = "../Project/Resources/Raw_Data/University_Enrollment.csv"
```

```python
# Read the Data file with the pandas library
uni_enrol_df = pd.read_csv(uni_enrol_file)

uni_enrol_df.head()
```

```python
organized_df = uni_enrol_df[['City', '2020-2021 (Undergrad)', '2020-2021 (Graduate)']]

organized_df.head()
```

```python
#Group by Cities and sum the categories
grouped_df = organized_df.groupby(['City']).sum()

grouped_df.head()
```

```python
#Sum the Undergrad + Graduate

total_enrollment = grouped_df['2020-2021 (Undergrad)'] + grouped_df['2020-2021 (Graduate)']

grouped_df['Total Enrollment'] = total_enrollment

grouped_df.sort_values(['Total Enrollment'], ascending=False)
```

```python
#Save the final copy to CSV

grouped_df.to_csv("../Project/Resources/Cleaned_Data/University_Enrollment.csv")
```
