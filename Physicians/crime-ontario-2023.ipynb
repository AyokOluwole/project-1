```python
# Dependencies
%matplotlib inline
import pandas as pd
import matplotlib.pyplot as plt
import scipy.stats as st
import numpy as np
from scipy.stats import linregress

```

```python
# Files to Load
ontarioCrime = "../Physicians/Resources/crime.csv"
population_data = "../Physicians/Resources/population.csv"
```

```python
#read ontariocrime File and store into Pandas DataFrames
totalCrime = pd.read_csv(ontarioCrime)
totalCrime.head()
```

```python
#read population File and store into Pandas DataFrames
population_data = "../Physicians/Resources/population.csv"
population = pd.read_csv(population_data)
population
```

```python
#merge population and totalcrime dataframe
final_data_with_population = pd.merge(totalCrime,population,how='left',on="Census subdivision name")
final_data_with_population
```

```python
#remove NaN to make clean data frame
clean_data = final_data_with_population.dropna()
clean_data.head()
```

```python
# assign axis for visualization
x = clean_data['Census subdivision name']
x
```

```python
y = clean_data['Total violation']
y
```

```python
z=clean_data['Total Population']
z
```

```python
#Bar chart showing Number of violation per city
plt.bar(x,y,align='center')

## Incorporate the other graph properties
plt.xlabel("Census subdivision name")
plt.ylabel("Number of violation")
plt.title("Number of Total violation per Census subdivision name")
plt.xticks(rotation="vertical")
plt.savefig("../Physicians/crime.png")
plt.show()
```

```python
#Scatter plot showing population vs Number of violation 
plt.scatter(y,z, edgecolor="black", linewidth=1, marker="o", alpha=1.0, label="Cities")

# Incorporate the other graph properties
plt.title("population Vs. Number of violation")
plt.ylabel("population(thousands)")
plt.xlabel("Number of violation")
plt.grid(True)
plt.xticks(rotation="vertical")
plt.savefig("../Physicians/crime2.png")
# Show plot
plt.show()
```

```python
#Compute Linear Regression for relationship between population and violation

def plot_linear_regression(x_values,y_values,title,text_coordinates):
    
    (slope, intercept, rvalue, pvalue, stderr) = linregress (x_values,y_values)
    regress_values = x_values * slope + intercept
    line_eq = "y = " + str(round(slope,2)) + " x + " + str(round(intercept,2))
    
    # Plot
    plt.scatter(x_values,y_values)
    plt.plot(x_values, regress_values, "r-")
    plt.annotate(line_eq,text_coordinates, fontsize=15, color="red")
    plt.xlabel("Total Population")
    plt.ylabel("Number of violation")
    print(f"The r-squared is: {rvalue}")
    plt.show()
```

```python
# Linear regression on Northern Hemisphere
x_values = clean_data ["Total Population"]
y_values = clean_data ["Total violation"]
plot_linear_regression(x_values,y_values, 'Total violation', (6,30))
plt.savefig("../Physicians/crime3.png")
plt.show()
```
